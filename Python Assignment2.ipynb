```python
n=5

for i in range(n):
    for j in range(i+1):
        print('* ',end='')
    print('\r')

for i in range(n-1,0,-1):
    for j in range(i):
        print('* ',end='')
    print('\r')
```

2. Write a Python program to reverse a word after accepting the input from the user.

Sample Output:

Input word: ineuron

Output: norueni

```python
string=input('Input word: ')
revrse_string=string[::-1]
print('Output: ' +revrse_string)
```
